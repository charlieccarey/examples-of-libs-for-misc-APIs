**We observe that some of this data looks messed up. The open high low for 12-20 and 12-21 look bad, far too low.**

```python
import quandl
```

```python
g_btc = quandl.get("GDAX/USD")
```

```python
g_btc.tail()
```

```python
g_btc.drop(['Volume'], axis=1).plot(figsize=[8,5])
```

```python
recent = quandl.get("GDAX/USD", start_date="2017-11-01", end_date="2017-12-22")
recent.drop(['Volume'], axis=1).plot(figsize=[8,5])
```

```python
recent['Volume'].plot(figsize=[8,5], legend='volume')
```

Let quandl handle the calculation of a rolling difference as a fraction (day vs. previous day difference as fraction change).

```python
change = quandl.get("GDAX/USD", start_date="2017-11-01", end_date="2017-12-22", transformation="rdiff")
```

As noted above, the data for 12-20 and 12-21 are suspect, leading to the large fraction changes on 12-22.

```python
change.tail()
```

```python
change.drop(['Volume'], axis=1).plot(figsize=[8,5])
```
